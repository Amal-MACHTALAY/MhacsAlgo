# Plotting the Rosenbrock function

In mathematical optimization, the Rosenbrock function is a non-convex function, introduced by Howard H. Rosenbrock in 1960, which is used as a performance test problem for optimization algorithms. It is also known as Rosenbrock's valley or Rosenbrock's banana function.

The global minimum is inside a long, narrow, parabolic shaped flat valley. To find the valley is trivial. To converge to the global minimum, however, is difficult.

## N-Dimension

$$J_{R}(v)=\sum_{i=1}^{i=N-1}\left\{\left(v_{i+1}-v_{i}^{2}\right)^{2}+\left(v_{i}-1\right)^{2}\right\}
$$


## 2-Dimension

The function is defined by :
$$ f(x,y)=(x-1)^{2}+(y-x^{2})^{2}$$
The Gradient :
$$ \nabla f(x,y)=\begin{pmatrix}
2(x-1)-4x(y-x^{2})\\ 
2(y-x^{2})
\end{pmatrix}$$
The Hessian :
$$H(f)_{(x,y)}= \begin{pmatrix}
2-4y+12x^{2} & -4x \\ 
-4x & 2
\end{pmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Rosenbrock(x,y):
    return (x-1)**2+(y-x**2)**2
```

```python
def Grad_Rosenbrock(x,y):
    g1=2*(x-1)-4*x*(y-x**2)
    g2=2*(y-x**2)
    return [g1,g2]
```

```python
def Hessian_Rosenbrock(x,y):
    h11=2-4*y+12*x**2
    h22=2
    h12=-4*x
    return [[h11,h12],[h12,h22]]
```

```python
x=np.arange(-2.0,2.0,0.1)
y=np.arange(-10.0,0.0,0.1)
X,Y=np.meshgrid(x,y)
Z=Rosenbrock(X,Y)
```

### Contour lines

```python
fig, ax =plt.subplots(figsize=(10,10))
ct=ax.contour(X,Y,Z, 30, cmap='brg')
plt.clabel(ct, inline=True, fontsize=8)
```

### 3D plot

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, 
                      cmap='jet',linewidth=1, antialiased=False)

ax.zaxis.set_major_locator(plt.LinearLocator(10))
ax.zaxis.set_major_formatter(plt.FormatStrFormatter('%.0f'))

#fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
